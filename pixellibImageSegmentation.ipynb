##Semantic Segmentation with PixelLib

### Importing necessary libraries

```python
import pixellib
from pixellib.semantic import semantic_segmentation
import matplotlib.pyplot as plt
```

### Importing trained model
You can available trained model on: https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5

```python
segment_image = semantic_segmentation()
segment_image.load_pascalvoc_model("/content/drive/MyDrive/pixellib/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5") 
```

### Image uploading

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

### Segmentation

```python
segment_image.segmentAsPascalvoc(fn, output_image_name = "image.jpg", overlay = True);
```

###Show image

```python
img = plt.imread("image.jpg")
plt.rcParams["figure.figsize"] = (20,20)
plt.imshow(img)
plt.show()
```
